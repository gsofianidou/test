```python
#εισαγωγή βιβλιοθηκών - πακέτων
import numpy as np
import pandas as pd
import seaborn as sns
import ppscore as pps
import matplotlib.pyplot as plt
```

```python
#εισαγωγή του συνόλου δεδομένων
dataset=pd.read_csv("https://media.githubusercontent.com/media/kdemertzis/EKPA/main/Data/Ransomware.csv",sep='|')
```

```python
#διαγραφή των μη κατηγορικών μεταβλητών
drop_elements = ['Name', 'md5']
dataset = dataset.drop(drop_elements, axis = 1)
```

```python
#υπολογισμός του πίνακα PPS μεταξύ όλων των στηλών
pps.matrix(dataset, sorted = True)
```

```python
#Οπτικοποίηση δεδομένων - bar plot
predictors_dataset = pps.predictors(dataset, y="SizeOfOptionalHeader")
plt.figure(figsize = (110,10))
sns.barplot(data=predictors_dataset, x="x", y="ppscore")
```

```python
#Οπτικοποίηση δεδομένων – heat map
new_var = pps.matrix
matrix_df = new_var(dataset)[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')
plt.figure(figsize = (55,55))
sns.heatmap(matrix_df, vmin=0, vmax=1, cmap="Blues", linewidths=0.5, annot=True)
```
